## Let's Create Our Final Web Scraper

```python
def get_title(soup):
    try:
        title = soup.find('span', attrs={'id':'productTitle'}).text.strip()
    except AttributeError:
        title = None
    return title
```

```python
def get_price(soup):
    try:
        price = soup.find('span', attrs={'id':'price'}).text.strip()
    except AttributeError:
        try:
            price = soup.find("span", attrs={'class':'a-price a-text-price header-price a-size-base a-text-normal'}).find('span', attrs={'class':'a-offscreen'}).text.strip()
        except:
            price = 'NONE'
    return price
```

```python
def get_rating(soup):
    try:
        rating = soup.find('i', attrs={'class':'a-icon a-icon-star a-star-4-5 cm-cr-review-stars-spacing-big'}).text.strip()
    except AttributeError:
        try:
            rating = soup.find("span", attrs={'class':'a-icon-alt'}).string.strip()
        except:
            rating = 'NONE'
    return rating
```

```python
def get_review_count(soup):
    try:
        review_count = soup.find('span', attrs={'id':'acrCustomerReviewText'}).text.strip()
    except AttributeError:
        review_count = 'NONE'
    return review_count
```

```python
def get_availabilty(soup):
    try:
        availability = soup.find('span', attrs={'class':'a-size-medium a-color-success'}).text.strip()
    except AttributeError:
        availability = 'NONE'
    return availability
```

```python
URL = 'https://www.amazon.com/s?k=data+science&crid=1LTLR6XBURI0W&sprefix=data+science%2Caps%2C320&ref=nb_sb_noss_2'

# Headers for request
HEADERS = {"User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/112.0.0.0 Safari/537.36", "Accept-Encoding":"gzip, deflate", "Accept":"text/html,application/xhtml+xml,application/xml;q=0.9,*/*;q=0.8", "DNT":"1","Connection":"close", "Upgrade-Insecure-Requests":"1"}

# HTTP request
webpage = requests.get(URL, headers = HEADERS)

# Creating Soup object
soup1 = bs4.BeautifulSoup(webpage.content, 'html.parser')

# Fetching the links as list of Tags
links = soup1.find_all('a', attrs = {'class' : "a-link-normal s-underline-text s-underline-link-text s-link-style a-text-normal"})

# Removing sponsored item links
for link in links:
    if 'https://' in link.get('href'):
        links.remove(link)
        
# Creating a List of links
link_list = []
for link in links:
    link_list.append('http://www.amazon.com'+link.get('href'))

# Creating a dictionary
dct = {'Title':[], 'Price':[],'Rating':[],'Review_Count':[],'Availability':[]}

# Extracting data from each product by Looping through Product Links
for link in link_list:
    
    inner_page = requests.get(link, headers = HEADERS)
    
    soup2 = bs4.BeautifulSoup(inner_page.content, 'html.parser')
    
    # Calling functions to load data in dictionary
    dct['Title'].append(get_title(soup2))
    dct['Price'].append(get_price(soup2))
    dct['Rating'].append(get_rating(soup2))
    dct['Review_Count'].append(get_review_count(soup2))
    dct['Availability'].append(get_availabilty(soup2))

# Creating DataFrame from Dictionary    
df = pd.DataFrame.from_dict(dct)
df = df.dropna(subset=['Title'])
df.head(50)
```
